```python
from bs4 import BeautifulSoup
import requests
import re

base_url = 'https://raw.githubusercontent.com'
urlpage = "https://github.com/CSSEGISandData/COVID-19/tree/master/csse_covid_19_data/csse_covid_19_daily_reports"
page = requests.get(urlpage)
soup = BeautifulSoup(page.content, 'html.parser')

table = soup.find('table', {'class':'files js-navigation-container js-active-navigation-container'})
results = table.find_all('tr')

for result in results[1:-1]:
    data = result.find_all('a', class_='js-navigation-open')
    
d1 = data[-1]
link = re.findall("href=[\"\'](.*?)[\"\']", str(d1))
url = base_url + ''.join(link).replace('blob/', '')
url
```

```python
import pandas as pd
df = pd.read_csv(url)
df
```

```python
import pandas as pd
df1 = pd.read_excel('forests.xlsx', skiprows=4).rename(columns={'Country Name': 'Country', 'Country Code': 'Code'})
df1 = df1.drop(df1.columns[3:34], axis=1)
df1 = df1.drop(df1.columns[-3:], axis=1)
df1
```

```python
df1 = df1.drop('Indicator Name', axis=1)
df1
```

```python
melt_df = pd.melt(df1,
                  id_vars = ['Country',
                             'Code'],
                  value_vars = df1.columns[2:].tolist(),
                  var_name = 'Date')

melt_df = melt_df.rename(columns={'value': 'Forest Area (sq. km)'})
melt_df
```

```python
import csv
melt_df.to_csv('forests.csv')
```
